```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ACLBSL_2000-01-01_2021-12-31.csv')
```

```python
data
```

```python
data.columns
```

```python
data['daily_return'] = (data['Close Price'] - data['Close Price'].shift(1)) / data['Close Price'].shift(1)
data.dropna(inplace=True)  # Drop rows with NaN returns 
```

```python
data
```

```python
daily_activity = data.groupby('Date')[['Total Transactions', 'Total Traded Shares', 'Total Traded Amount']].sum()
```

```python
data
```

```python
print("Daily Activity:")
print(daily_activity)
```

```python
print("Daily Activity:")
print(daily_activity)

# Price Analysis (assuming numeric conversion successful)
daily_prices = data.groupby('Date')[['Max. Price', 'Min. Price', 'Close Price']].describe()
print("\nDaily Price Analysis:")
print(daily_prices)

# Price Volatility (assuming numeric conversion successful)
data['Price Range'] = data['Max. Price'] - data['Min. Price']
daily_volatility = data.groupby('Date')['Price Range'].mean()
print("\nDaily Price Volatility (Average Daily Range):")
print(daily_volatility)
```
